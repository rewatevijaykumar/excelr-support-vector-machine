# Assignment-17-SVM-Forest Fires

### Classify the Size_Categorie using SVM

month	month of the year: 'jan' to 'dec'\
day	day of the week: 'mon' to 'sun'\
FFMC	FFMC index from the FWI system: 18.7 to 96.20\
DMC	DMC index from the FWI system: 1.1 to 291.3\
DC	DC index from the FWI system: 7.9 to 860.6\
ISI	ISI index from the FWI system: 0.0 to 56.10\
temp	temperature in Celsius degrees: 2.2 to 33.30\
RH	relative humidity in %: 15.0 to 100\
wind	wind speed in km/h: 0.40 to 9.40\
rain	outside rain in mm/m2 : 0.0 to 6.4\
Size_Categorie 	the burned area of the forest ( Small , Large)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# import dataset
df = pd.read_csv('dataset/forestfires.csv')
df.head()
```

```python
df.info()
```

```python
#already encoded values for month and day in dataset
df.drop(['month','day'], axis=1, inplace=True)
df.head()
```

```python
# Label Encoding
df['size_category'] = LabelEncoder().fit_transform(df['size_category'])
df.head()
```

```python
# split dataset into features and target
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Standardize values
X = StandardScaler().fit_transform(X)
```

```python
# split into train and test
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# Support Vector Classifier
clf = SVC()
#Parameter Selection for SVM using GridSearchCV
param_grid = [{'kernel':['rbf','linear','poly'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
#pass the optimal parameters to Classifier and check the score
clf = SVC(kernel="linear",C= 15, gamma = 50)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_pred,y_test))
```
